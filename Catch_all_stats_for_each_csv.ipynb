# **Bureau CSV**

All client's previous credits provided by other financial institutions that were reported to Credit Bureau (for clients who have a loan in our sample).

For every loan in our sample, there are as many rows as number of credits the client had in Credit Bureau before the application date.

These columns collectively provide a detailed view of a client's previous and ongoing credit obligations.

## **Important Notes -**

Currency 1: The account is "current" and up-to-date with payments. This means that there are no missed payments and the borrower is in good standing.

Currency 2: The account is one payment cycle past due. This typically means the payment is 30 days late.

Currency 3: The account is two payment cycles past due, usually indicating that the payment is 60 days late.
